```python

def calculate_competitive_advantage(facebook_df, ins_df, num_variables):

    competitive_advantage = {}

    # Variable names as given in the problem statement
    variable_names = ['정보', '오락', '개인화', '상호작용', '만족']

    # Start index for each variable's measurements in the DataFrame
    start_index = 0

    # Loop through each variable by its number of measurements
    for var_index, measurements in enumerate(num_variables):
        # Calculate the sum of scores where media 1 has a higher score than media 2
        M_i_greater_j = facebook_df.iloc[:, start_index:start_index + measurements].gt(
                            ins_df.iloc[:, start_index:start_index + measurements]).multiply(
                            facebook_df.iloc[:, start_index:start_index + measurements]).sum().sum()
        # Calculate the sum of scores where media 2 has a higher score than media 1
        M_j_greater_i = ins_df.iloc[:, start_index:start_index + measurements].gt(
                            facebook_df.iloc[:, start_index:start_index + measurements]).multiply(
                            ins_df.iloc[:, start_index:start_index + measurements]).sum().sum()

        # Calculate competitive advantage scores and add to dictionary
        competitive_advantage[variable_names[var_index]] = {
            'M_i_greater_j': M_i_greater_j,
            'M_j_greater_i': M_j_greater_i,
            'S_i_greater_j': M_i_greater_j / len(facebook_df),
            'S_j_greater_i': M_j_greater_i / len(ins_df)
        }

        # Update the start index for the next variable
        start_index += measurements

    return competitive_advantage

# Number of measurements for each variable in the order: 정보, 오락, 상호작용, 개인화, 만족
num_variables = [5, 4, 3, 4, 4]

# Calculate competitive advantage between Facebook and INS
comp_adv_fb_ins = calculate_competitive_advantage(facebook_df, ins_df, num_variables)

comp_adv_fb_ins
```

```python

def calculate_variable_averages(df, num_measurements):
    averages = {}
    variable_names = ['정보', '오락', '개인화', '상호작용', '만족']
    start_index = 0

    # calculate the average
    for i, (variable, num) in enumerate(zip(variable_names, num_measurements)):

        columns = df.iloc[:, start_index:start_index + num]

        averages[variable] = columns.mean().mean()

        start_index += num

    return averages
```
